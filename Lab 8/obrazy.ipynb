```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.spatial.distance import pdist, squareform
```

```python
data = np.loadtxt("data.dat", dtype=float, delimiter=" ")
print(data)
```

```python
X = data[:,0]
Y = data[:,1]
Z = data[:,2]


fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Plot the surface
ax.plot(X, Y, Z, "ko")

# Set an equal aspect ratio
ax.set_aspect('equal')

plt.show()

```

```python
points = data

# Obliczenie odlegoci midzy wszystkimi parami punktw
distances = squareform(pdist(points))

# Warto maksymalna, ktra bdzie uznana za wystarczajco blisk
max_distance = 1.8  # Tutaj moesz ustawi odpowiedni warto

# Tworzenie listy przechowujcej poczenia midzy punktami
connections = []

# Szukanie pocze
for i in range(len(points)):
    for j in range(i+1, len(points)):
        if distances[i, j] <= max_distance:
            connections.append((points[i], points[j]))

# Tworzenie wykresu 3D
fig = plt.figure(figsize=(4,4), dpi=200)
ax = fig.add_subplot(111, projection='3d')

# Dodawanie punktw
ax.scatter(points[:,0], points[:,1], points[:,2], color="red")

# Dodawanie linii czcych punkty
for connection in connections:
    ax.plot3D(*zip(*connection), color='black', linewidth=1.5, alpha=0.8)

# Ustawienie etykiet osi
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Wywietlenie wykresu
plt.show()
```
