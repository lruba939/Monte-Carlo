```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
relaksacja = np.loadtxt("V_MC.dat", dtype=float, delimiter=" ", usecols=(range(0,31,1)))
X = range(0, 31, 1)
Y = range(0, 31, 1)
plt.pcolormesh(X, Y, relaksacja, cmap="hot")
plt.colorbar()
plt.title(r'V$_{\text{MC}}$')
plt.xlabel(r'i')
plt.ylabel(r'j')
```

```python
relaksacja = np.loadtxt("V_relaksacja.dat", dtype=float, delimiter=" ", usecols=(range(0,31,1)))
MC = np.loadtxt("V_MC.dat", dtype=float, delimiter=" ", usecols=(range(0,31,1)))
data = abs(relaksacja-MC)
X = range(0, 31, 1)
Y = range(0, 31, 1)
plt.pcolormesh(X, Y, data, cmap="hot")
plt.colorbar()
plt.title(r'|V$_{\text{MC}}$ - V$_{\text{rel}}$|')
plt.xlabel(r'i')
plt.ylabel(r'j')
```
