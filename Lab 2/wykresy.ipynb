```python
import numpy as np
# import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
def dystrybuanta(x_k, x_0):
    prop = 4/5*( (x_k + x_k**2/2 - x_k**4/4) - (x_0 + x_0**2/2 - x_0**4/4) )
    return prop
```

```python
zlozony = np.loadtxt("output_zlozony.dat", comments="#", delimiter=" ", unpack=False)
zlozony = np.asarray(zlozony[:,1])

x = np.linspace(0, 1, 101)
f = 0.8 * (1 + x - x**3)

plt.figure(figsize=(9,6), dpi=300)
plt.hist(zlozony, density=True, label="Algorytm rozkładu złożonego")
plt.plot(x, f, 'r', label="Funkcja gęstości prawdopodobieństwa")
plt.xticks(np.arange(0, 1, step=0.1)) 
plt.grid(True)
plt.ylim([0.8, 1.2])
plt.legend(loc="upper left")
plt.xlabel("Zmienna losowa U")
plt.ylabel("Znormalizowana liczba zliczeń")

histogram_data = np.histogram(zlozony, np.arange(0,1+0.1,0.1))
n = np.array(histogram_data[0])
bins = np.array(histogram_data[1])
# print(bins)
N = 10**6

chi2 = 0
for i in range(10):
    p_i = dystrybuanta(bins[i+1], bins[i])
    chi2 += (n[i] - p_i*N)**2/(p_i*N)
print(chi2)
```
