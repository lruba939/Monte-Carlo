```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import matplotlib.patches as patches
```

```python
hist = np.loadtxt("histogram.dat", dtype=float)
x = np.arange(0, 8, 8/200)
psi_100 = 4*x**2*np.exp(-2*x)

textsize = 12

plt.figure(figsize=([6,4]), dpi=300)
ax = plt.subplot()

ax.plot(x, psi_100, "--k", label = "Teoria")
ax.plot(x, hist, color="tab:orange", label = "Monte Carlo")

ax.set_xlabel(r"r [a$_b$]", fontsize = textsize)

ax.tick_params(direction="in", which="both", top=True, right=True)
ax.grid(True, which="both", alpha=0.2)
ax.xaxis.set_minor_locator(MultipleLocator(2))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
ax.set_xlim([0,8])
ax.set_ylim([0,0.6])
ax.legend(frameon=False, fontsize = textsize)
```

```python
def plot_maps(data, title, mapa, ax, fig):
    textsize = 18

    X = range(0, len(data[1,:]), 1)
    Y = range(0, len(data[:,1]), 1)
    x = np.linspace(0.3, 1.2, num=45)
    y = np.linspace(-0.7, 0.3, num=50)
    im = ax.pcolormesh(x, y, data, cmap=mapa)
    fig.colorbar(im, ax=ax)
    ax.set_title(title, fontsize = textsize)
    ax.set_xlabel(r'a', fontsize = textsize)

    if ax is axs[0]:
        ax.set_ylabel(r'c', fontsize = textsize)
```

```python
fig, axs = plt.subplots(1, 3, figsize=([22,5]), dpi=600)

energy = np.loadtxt("D:\AGH\Semestr 8\Monte-Carlo\Lab 14\map_en.dat", dtype=float)
title_en = r"Energy"
plot_maps(energy, title_en, "hot", axs[0], fig)

std = np.loadtxt("D:\AGH\Semestr 8\Monte-Carlo\Lab 14\map_std.dat", dtype=float)
title_std = r"$\sigma$"
plot_maps(std, title_std, "viridis", axs[1], fig)

log_data = np.log10(std + 10**(-20))
title_log = r"log($\sigma + 10^{-20}$)"
plot_maps(log_data, title_log, "cividis", axs[2], fig)
```
