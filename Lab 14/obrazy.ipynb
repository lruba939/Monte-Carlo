```python
def plot_maps(data, title):
    X = range(0, len(data[1,:]), 1)
    Y = range(0, len(data[:,1]), 1)
    x = np.linspace(0.3, 1.2, num=45)
    y = np.linspace(-0.7, 0.3, num=50)
    plt.pcolormesh(x, y, data, cmap="hot")
    # plt.xticks(np.arange(0.3, 1.2, 0.02))
    # plt.yticks(np.arange(-0.7, 0.3, 0.02))
    plt.colorbar()
    plt.title(title)
    plt.xlabel(r'a')
    plt.ylabel(r'c')
```

```python
energy = np.loadtxt("D:\AGH\Semestr 8\Monte-Carlo\Lab 14\map_en.dat", dtype=float)
title_en = r"Energy"
plot_maps(energy, title_en)
plt.show()

std = np.loadtxt("D:\AGH\Semestr 8\Monte-Carlo\Lab 14\map_std.dat", dtype=float)
title_std = r"$\sigma$"
plot_maps(std, title_std)
plt.show()

log_data = np.log10(std + 10**(-20))
title_log = r"log($\sigma + 10^{-20}$)"
plot_maps(log_data, title_log)
plt.show()
```
