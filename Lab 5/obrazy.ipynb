```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(-3, 3, 0.01)
Y = 1+np.tanh(X)

plt.figure(figsize=(6,4), dpi=300)
plt.plot(X, Y, 'r-', label="C1")
plt.grid('minor')
plt.legend()
```

```python
X = np.arange(0, 10, 0.01)
Y = 1/(1+X**2)

plt.figure(figsize=(6,4), dpi=300)
plt.plot(X, Y, 'r-', label="C2")
plt.grid('minor')
plt.legend()
```

```python
X = np.arange(0, 1, 0.001)
Y = np.cos(np.pi*X)**(10)

plt.figure(figsize=(6,4), dpi=300)
plt.plot(X, Y, 'r-', label="C3")
plt.grid('minor')
plt.legend()
```

```python
log = np.loadtxt("log.dat", comments="#", delimiter=" ", unpack=False)

X = np.arange(-3, 3, 0.01)

# dla 10^2 losowa
log_c1_1 = log[0:101]
log_c1_2 = log[101:201]
log_c1_3 = log[201:308]

biny = np.arange(-3,3+0.6,0.6)

Y = 1+np.tanh(X)
con = np.histogram(log_c1_3, biny)
Y = Y/np.max(Y)*np.max(con[0])

# print(np.max(con[0]))

color1 = "#bf4713"
color2 = "#845b43"
color3 = "#e90046"

plt.figure(figsize=(6,4), dpi=300)

plt.hist(log_c1_1, bins=biny, histtype='step', linewidth=2.5, alpha = 1, label="Metoda podstawowa")
plt.hist(log_c1_2, bins=biny, histtype='step', linewidth=2.5, alpha = 1, label="Metoda warstwowa nieoptymalna")
plt.hist(log_c1_3, bins=biny, histtype='step', linewidth=2.5, alpha = 1, label="Metoda warstwowa optymalna")

plt.xticks(biny)
plt.title("$10^2$ losowa")
plt.xlabel("Przedzia wartoci zmiennej losowej x [-]")
plt.ylabel("Ilo losowa w danym przedziale [-]")
plt.legend(fontsize=8)

plt.plot(X, Y, 'r-', alpha=1)

plt.grid(True)
```

```python
data = np.loadtxt("podst.dat", comments="#", delimiter=" ", unpack=False)
data2 = np.loadtxt("sys.dat", comments="#", delimiter=" ", unpack=False)
data3 = np.loadtxt("opt.dat", comments="#", delimiter=" ", unpack=False)

x = [10**2, 10**3, 10**4, 10**5]

color1 = "#877ab7"
color2 = "#7ab787"
color3 = "#fcba03"

plt.figure(figsize=(6,4), dpi=300)

plt.errorbar(x, data[0:4,1], data[0:4,4], marker="o", mfc=color1, linestyle="None", mec="k", ecolor=color1, capthick=True, capsize=4, label="Metoda podstawowa")
plt.errorbar(x, data2[0:4,1], data2[0:4,4], marker="o", mfc=color2, linestyle="None", mec="k", ecolor=color2, capthick=True, capsize=4, label="Metoda warstwowa nieoptymalna")
plt.errorbar(x, data3[0:4,1], data3[0:4,4], marker="o", mfc=color3, linestyle="None", mec="k", ecolor=color3, capthick=True, capsize=4, label="Metoda warstwowa optymalna")
plt.axhline(y = 6, color = 'r', linestyle = '--', alpha=0.75)

plt.xscale("log")
plt.grid(True)
plt.title(r'$C_1 = \int_{-3}^{3}(1+tanh(x))dx$')
plt.xlabel("Ilo wygenerowanych punktw [-]")
plt.ylabel("Warto caki [-]")
plt.legend(fontsize = 8)
```
