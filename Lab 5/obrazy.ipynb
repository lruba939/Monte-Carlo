```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(0, 1000, 1)
Y = np.sqrt(X)

plt.figure(figsize=(6,4), dpi=300)
plt.plot(X, Y, 'r-', label="TEST")
plt.grid('minor')
plt.legend()
```

```python
data = np.loadtxt("podst.dat", comments="#", delimiter=" ", unpack=False)
data2 = np.loadtxt("sys.dat", comments="#", delimiter=" ", unpack=False)

x = [10**2, 10**3, 10**4, 10**5]

color1 = "#877ab7"
color2 = "#7ab787"

plt.figure(figsize=(6,4), dpi=300)

plt.errorbar(x, data[0:4,2], data[0:4,5], marker="o", mfc=color1, linestyle="None", mec="k", ecolor=color1, capthick=True, capsize=4)
plt.errorbar(x, data2[0:4,2], data2[0:4,5], marker="o", mfc=color2, linestyle="None", mec="k", ecolor=color2, capthick=True, capsize=4)

plt.xscale("log")
plt.grid(True)
plt.title("C1")
plt.xlabel("Ilość wygenerowanych punktów [-]")
plt.ylabel("Wartość całki [-]")
```

```python
plt.figure(figsize=(6,4), dpi=300)

plt.errorbar(x, data[4:8,2], data[4:8,5], marker="o", mfc=color1, linestyle="None", mec="k", ecolor=color1, capthick=True, capsize=4)
plt.errorbar(x, data2[4:8,2], data2[4:8,5], marker="o", mfc=color2, linestyle="None", mec="k", ecolor=color2, capthick=True, capsize=4)

plt.xscale("log")
plt.grid(True)
plt.title("C2")
plt.xlabel("Ilość wygenerowanych punktów [-]")
plt.ylabel("Wartość całki [-]")
```
