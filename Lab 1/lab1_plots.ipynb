```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('output.dat', sep = " ", header=None)
data = data.to_numpy()

plt.figure(1)
plt.loglog(data[1:7,0], data[1:7,4], "s--", label = "p = 0.1")
plt.loglog(data[8:14,0], data[8:14,4], "s--", label = "p = 0.5")
plt.loglog(data[15:21,0], data[15:21,4], "s--", label = "p = 0.7")
plt.grid(True)
plt.xlabel("Ilość elementów ciągu n [-]")
plt.ylabel("Błąd względny wartości średniej <X(n)> [-]")
plt.legend()
```

```python
plt.figure(2)
plt.loglog(data[1:7,0], data[1:7,5], "s--", label = "p = 0.1")
plt.loglog(data[8:14,0], data[8:14,5], "s--", label = "p = 0.5")
plt.loglog(data[15:21,0], data[15:21,5], "s--", label = "p = 0.7")
plt.grid(True)
plt.xlabel("Ilość elementów ciągu n [-]")
plt.ylabel("Błąd względny wariancji <X(n)> [-]")
plt.legend()
```
