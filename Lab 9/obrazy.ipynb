```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
relaksacja = np.loadtxt("V_relaksacja.dat", dtype=float, delimiter=" ", usecols=(range(0,31,1)))
X = range(0, 31, 1)
Y = range(0, 31, 1)
plt.pcolormesh(X, Y, relaksacja, cmap="hot")
plt.colorbar()
plt.title(r'V$_{\text{rel}}$')
plt.xlabel(r'i')
plt.ylabel(r'j')
```

```python
data1 = np.loadtxt("V_MC_1.dat", dtype=float, delimiter=" ", usecols=(range(0,31,1)))
data2 = np.loadtxt("V_MC_2.dat", dtype=float, delimiter=" ", usecols=(range(0,31,1)))
data3 = np.loadtxt("V_MC_3.dat", dtype=float, delimiter=" ", usecols=(range(0,31,1)))

X = range(0, 31, 1)
Y = range(0, 31, 1)

plt.figure(figsize=(20,5))
########################
plt.subplot(1,3,1)
plt.pcolormesh(X, Y, data1, cmap="hot")
plt.colorbar()
plt.title(r'N$_{\text{chains}}$ = 100, n$_{\text{length}}$ = 100, B$_{j_0,i_0}$ = 0')
plt.xlabel(r'i')
plt.ylabel(r'j')
########################
plt.subplot(1,3,2)
plt.pcolormesh(X, Y, data2, cmap="hot")
plt.colorbar()
plt.title(r'N$_{\text{chains}}$ = 100, n$_{\text{length}}$ = 100, B$_{j_0,i_0}$ = 1')
plt.xlabel(r'i')
plt.ylabel(r'j')
########################
plt.subplot(1,3,3)
plt.pcolormesh(X, Y, data3, cmap="hot")
plt.colorbar()
plt.title(r'N$_{\text{chains}}$ = 300, n$_{\text{length}}$ = 300, B$_{j_0,i_0}$ = 1')
plt.xlabel(r'i')
plt.ylabel(r'j')

plt.suptitle(r'Rozkład potencjału V obliczonego metodą Monte Carlo')
plt.subplots_adjust(top=0.8)
```

```python
relaksacja = np.loadtxt("V_relaksacja.dat", dtype=float, delimiter=" ", usecols=(range(0,31,1)))
MC = np.loadtxt("V_MC.dat", dtype=float, delimiter=" ", usecols=(range(0,31,1)))
data = abs(relaksacja-MC)
X = range(0, 31, 1)
Y = range(0, 31, 1)
plt.pcolormesh(X, Y, data, cmap="hot")
plt.colorbar()
plt.title(r'|V$_{\text{MC}}$ - V$_{\text{rel}}$|')
plt.xlabel(r'i')
plt.ylabel(r'j')
```

```python
data1 = np.loadtxt("V_MC_1.dat", dtype=float, delimiter=" ", usecols=(range(0,31,1)))
data2 = np.loadtxt("V_MC_2.dat", dtype=float, delimiter=" ", usecols=(range(0,31,1)))
data3 = np.loadtxt("V_MC_3.dat", dtype=float, delimiter=" ", usecols=(range(0,31,1)))

data1 = np.abs(data1 - relaksacja)
data2 = np.abs(data2 - relaksacja)
data3 = np.abs(data3 - relaksacja)

X = range(0, 31, 1)
Y = range(0, 31, 1)

plt.figure(figsize=(20,5))
########################
plt.subplot(1,3,1)
plt.pcolormesh(X, Y, data1, cmap="hot")
plt.colorbar()
plt.title(r'N$_{\text{chains}}$ = 100, n$_{\text{length}}$ = 100, B$_{j_0,i_0}$ = 0')
plt.xlabel(r'i')
plt.ylabel(r'j')
########################
plt.subplot(1,3,2)
plt.pcolormesh(X, Y, data2, cmap="hot")
plt.colorbar()
plt.title(r'N$_{\text{chains}}$ = 100, n$_{\text{length}}$ = 100, B$_{j_0,i_0}$ = 1')
plt.xlabel(r'i')
plt.ylabel(r'j')
########################
plt.subplot(1,3,3)
plt.pcolormesh(X, Y, data3, cmap="hot")
plt.colorbar()
plt.title(r'N$_{\text{chains}}$ = 300, n$_{\text{length}}$ = 300, B$_{j_0,i_0}$ = 1')
plt.xlabel(r'i')
plt.ylabel(r'j')

plt.suptitle(r'Rozkład różnicy potencjału V$_{MC}$ i V$_{rel}$')
plt.subplots_adjust(top=0.8)
```
