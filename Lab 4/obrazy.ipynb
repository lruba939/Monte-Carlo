```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(0, 1000, 1)
Y = np.sqrt(X)

plt.figure(figsize=(6,4), dpi=300)
plt.plot(X, Y, 'r-', label="TEST")
plt.grid('minor')
plt.legend()
```

```python
KA = np.loadtxt("KA.dat", comments="#", delimiter=" ", unpack=False)
KB = np.loadtxt("KB.dat", comments="#", delimiter=" ", unpack=False)

colorA = "#b7a57a"
colorB = "#7a8cb7"

plt.figure(figsize=(6.5,4), dpi=300)
plt.plot(KA[:,0], KA[:,1], 'o', markeredgecolor=colorA, markerfacecolor=colorA, markersize=0.6, alpha=1, label="Koo o parametrach A")
plt.plot(KB[:,0], KB[:,1], 'o', markeredgecolor=colorB, markerfacecolor=colorB, markersize=0.6, alpha=1, label="Koo o parametrach B")
plt.grid(True)
plt.xlabel("Wsprzdna X")
plt.ylabel("Wsprzdna Y")
plt.legend(markerscale=8, fontsize=8)
```

```python
a = np.loadtxt("3a.dat", comments="#", delimiter=" ", unpack=False)
c = np.loadtxt("3c.dat", comments="#", delimiter=" ", unpack=False)

x = [10**2, 10**3, 10**4, 10**5, 10**6]

color1 = "#877ab7"
color2 = "#7ab787"

plt.figure(figsize=(6,4), dpi=300)

plt.errorbar(x, a[:,1], a[:,4], marker="o", mfc=color1, linestyle="None", mec="k", ecolor=color1, capthick=True, capsize=4, label=' = A')
plt.errorbar(x, c[:,1], c[:,4], marker="o", mfc=color2, linestyle="None", mec="k", ecolor=color2, capthick=True, capsize=4, label=' = B')


plt.xscale("log")
plt.grid(True)
plt.legend()
plt.title("$x_a=R_b+0.5\cdot R_a$")
plt.xlabel("Ilo wygenerowanych punktw [-]")
plt.ylabel("Cz wsplna [-]")
```
